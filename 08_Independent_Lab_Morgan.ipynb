# Rename Columns

```python
# Rename columns as directed and save as new DataFrame to preserve history.
gh_tfp_renamed = (gh_turnover_four_projects
        .rename(columns={'domain':'project_domain', 
                         'language':'project_language',
                         'windows':'project_age',
                         'window_idx':'quarter',
                         'num_team':'team_size',
                         'num_commits':'project_commits',
                         'blau_gender':'gender_blau',
                         'Gini_gh_ten':'gh_tenure_gini',
                         'Core1':'core_dev',
                         'commits':'user_commits',
                         'propCommits':'commit_proportion',
                         'github_tenure':'gh_tenure',
                         'leavesNextQ':'leaves_next_q'})
)

# Quick inspection:
gh_tfp_renamed
```

# Data Wrangling

## Calculate the mean of `user_commits`. 

```python
# Calculate by dividing the sum of 'user_commits' by the number of rows in the DataFrame.
sum(gh_tfp_renamed.user_commits) / len(gh_tfp_renamed)
```

```python
# Or! Use the NumPy mean() function. This is much easier. See reference. 
np.mean(gh_tfp_renamed.user_commits)
```

## Calculate the median of `user_commits`.

```python
# Calculate the median using the NumPy median() function. Neat!
np.median(gh_tfp_renamed.user_commits)
```

## Select all columns that start with a *p* (i.e., `(^p)`) **or** contain an *g* (i.e., `(g)`). Save it as a new DataFrame named `turnover_new`. Output the columns in your notebook.

```python
# Select columns as instructed using .filter() and save as new DataFrame titled turnover_new.
turnover_new = (gh_tfp_renamed
 .filter(regex = '(^p)|(g)')
)

# Output the columns.
turnover_new.columns
```

## Using your newly created DataFrame `turnover_new`, select rows in which `project_commits` is greater than 9,000.


```python
# Use the .query() function to get project_commits greater than 9000.
(turnover_new
 .query("project_commits > 9000")
)
```

## Using pandas piping notation, perform the previous two operations together and save it as a new DataFrame `turnover_newer`. This means you should select columns that start with a *p* or contain an *g* and select rows in which `project_commits` is greater than 9,000.

```python
# Combine the previous two operations and save as turnover_newer.
turnover_newer = (gh_tfp_renamed
 .filter(regex = '(^p)|(g)')
 .query("project_commits > 9000")
)

# Check our work. They seem to be the same. 
turnover_newer
```
